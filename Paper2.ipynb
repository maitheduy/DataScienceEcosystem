### 1. Introduction
<p align="justify">Hausdorff dimension is a fundamental concept in fractal geometry. It is a way to quantify the "roughness" or "complexity" of a fractal set. Marstrand’s projection theorem (1954) [7] states that for any set $ F $ in the Euclidean plane with a Hausdorff dimension greater than 1, the orthogonal projection of $ F $ onto almost every line $ L $ will have positive Lebesgue measure. In a more recent development, Rams and Simon [6] proved that the concept of "having a positive Lebesgue measure" can be elevated to "encompassing an interval" for Cantor dust sets. In this research paper, we delve into the inquiry of when "almost every" can be enhanced to "each and every," and whether "encompassing an interval" can be heightened to "constituting an interval."

In [1], Dijkstra et al. proved that there exists a planar Cantor set of vanishing Lebesgue measure with the property that every projection is an interval. Building upon their work, this paper further extends their concept by establishing that the infimum Hausdorff dimension of such sets is equal to 1. It is known that the projection of irregular 1-set $ F $ has length zero for almost all lines $ L $ [2]. However, it remains an open question whether 1 represents the minimum attainable value.</p>

### 2. Preliminaries

<p align="justify"> At first, we present some basic definitions and facts as the properties of the Hausdorff dimension of sets and delve into the Hausdorff dimension of self-similar sets.

A fundamental result, as established by Hutchinson in [3], is that for any integer $n \geq 2$ and a set of contraction similarity maps $\{f_1, \dots, f_n\}$ on $\mathbb{R}^d$, there exists a unique compact set $F$ that satisfies the equation $\displaystyle \bigcup_{i=1}^n f_i(F) = F$. This set of contraction similarity maps $\{f_1, f_2, \dots, f_n\}$ is referred to as the iterated function system (is abbreviated as IFS) and the set $F$ is known as a self-similar set induced by the IFS $\{f_1, \dots, f_n\}$.

Given a subset $F$ of $\mathbb{R}^d$, let $\dim_H(F)$ be the Hausdorff dimension of $F$. According to [2], when the Hausdorff dimension of $F$ is too small, the set $F$ is totally disconnected.

Calculating the Hausdorff dimension of a set is quite difficult. The following is the simplest case in which we can calculate the Hausdorff dimension.

> **Definition 2.1.**  
> <p align="justify"><em>We said that an IFS $\{f_1, f_2, \dots, f_n\}$ on $\mathbb{R}^d$ is satisfied the open set condition if there exists a non-empty bounded open set $V$ in $\mathbb{R}^d$ such that
> $$ \bigcup_{i=1}^n f_i(V) \subset V \text{ and } f_i(V) \cap f_j(V) = \emptyset \quad (i \neq j), \quad i, j = 1, \dots, n. $$
> When this condition is met, the Hausdorff dimension of a self-similar set can be accurately computed, as elaborated in [2] and [5].</em></p>

---

> **Theorem 2.2.**  
> <p align="justify"><em>Suppose that the open set condition holds for the contraction similarity maps $\{f_1, \dots, f_n\}$ with ratios $0 < r_i < 1$, $i = 1, \dots, n$ on $\mathbb{R}^d$. If $F$ is the self-similar set of IFS $\{f_1, \dots, f_n\}$, then $\dim_H(F) = s$ where $s$ is a unique solution of the equation
> $$ \sum_{i=1}^n r_i^s = 1. $$</em></p>

---

<p align="justify">One of the applications of the Hausdorff dimension in topology is demonstrated in the following result (we can see it in [2]).</p>

> **Proposition 2.3.**  
> <p align="justify"><em>Let $F$ be a subset of $\mathbb{R}^d$. If $\dim_H F < 1$ then $F$ is totally disconnected.</em></p>

<p align="justify">According to [2], the relationship between the Hausdorff dimension of a set and that of its image under a Lipschitz map is as follows.</p>

> **Proposition 2.4.**  
> <p align="justify"><em>If $f : F \to \mathbb{R}^m$ is a Lipschitz transformation then
> $$ \dim_H f(F) \leq \dim_H(F). $$</em></p>

<p align="justify">Now, we will state and provide a proof for the main result.</p>

### 3. Main Results

Let $F$ be a class of totally disconnected subsets $\mathbb{F}$ of $\mathbb{R}^2$ such that the orthogonal projection of $F$ onto any line is an interval. At first, we have the following results.

> **Theorem 3.1.** _Let $F$ be an element of the collection $\mathbb{F}$. Then, the Hausdorff dimension of $F$, $\dim_H F$, is greater than or equal to 1._

**Proof.** Assume the contrary, let $F$ be a set with a Hausdorff dimension strictly less than 1. Let $p$ be the projection from $\mathbb{R}^2$ onto $\{(x, 0) : x \in \mathbb{R}\}$ axis. It is evident that $p$ is a Lipschitz map. Consequently, as per Proposition 2.4, $\dim_H(p(F)) \leq \dim_H(F) < 1$. Therefore, according to Proposition 2.3, $p(F)$ is totally disconnected. As a result, $p(F)$ cannot possibly constitute an interval. This contradicts the assumption. Therefore, the Hausdorff dimension of $F$ is greater than or equal to 1. □

> **Theorem 3.2.** _With the collection $\mathbb{F}$ defined as above, we have_
> $$ \inf \{\dim_H F : F \in \mathbb{F}\} = 1. $$

**Proof.** We will show that, let $s$ be a value greater than 1 and arbitrarily close to 1; there exists a self-similar set $A \in \mathbb{F}$ with $\dim_H(A) = s$, meaning $A$ is totally disconnected and the projection of $A$ onto any line is an interval.

In fact, given $s$ as above, there exists a unique $\lambda$ in the interval $\left(0, \frac{\sqrt{5} - 1}{2}\right)$ such that
$$ 3\lambda^s + (1 - \lambda - \lambda^2)^s = 1. $$

Now, we consider an iterated function system comprising the following four functions:
$$
f_1(x) = \lambda R_1 x + (0, 1), \\
f_2(x) = \lambda x, \\
f_3(x) = \lambda R_3 x + (1, 0), \\
f_4(x) = (1 - \lambda - \lambda^2)x + \left(\lambda/2 + \lambda^2/2, \lambda/2 + \lambda^2/2\right),
$$
where $R_1$ and $R_3$ are matrices representing reflections across the x-axis and y-axis, respectively. It is clear that $f_1, \dots, f_4$ are similar maps with similarity ratios $r_1 = r_2 = r_3 = \lambda$, $r_4 = 1 - \lambda - \lambda^2$. So, let $A$ be the corresponding self-similar set of this IFS.


```python
# @title Trường hợp $\lambda=1/3$
import matplotlib.pyplot as plt
import numpy as np

# Set the lambda value
lambda_val = 0.33333333

# Define the transformation matrices
R1 = np.array([[1, 0], [0, -1]])  # Symmetry about the x-axis (Ox)
R3 = np.array([[-1, 0], [0, 1]])  # Symmetry about the y-axis (Oy)

# Define the transformation functions based on the given formulas
def f1(point):
    return lambda_val * R1 @ point + np.array([0, 1])

def f2(point):
    return lambda_val * point

def f3(point):
    return lambda_val * R3 @ point + np.array([1, 0])

def f4(point):
    return (1 - lambda_val - lambda_val**2) * point + np.array([
        (lambda_val / 2 + lambda_val**2 / 2),
        (lambda_val / 2 + lambda_val**2 / 2)
    ])

# Chaos game to draw the attractor using given transformations and probabilities
def custom_chaos_game_colored(iterations):
    point = np.array([0.5, 0.5])  # Start with an initial point
    points_f1 = []
    points_f2 = []
    points_f3 = []
    points_f4 = []

    for _ in range(iterations):
        r = np.random.rand()
        if r < 0.2:
            point = f1(point)  # Probability 0.2
            points_f1.append(point)
        elif r < 0.4:
            point = f3(point)  # Probability 0.2
            points_f3.append(point)
        elif r < 0.7:
            point = f2(point)  # Probability 0.3
            points_f2.append(point)
        else:
            point = f4(point)  # Probability 0.3
            points_f4.append(point)

    return (np.array(points_f1), np.array(points_f2),
            np.array(points_f3), np.array(points_f4))

# Set the number of iterations
iterations = 500000

# Generate the attractor using the custom Chaos Game
points_f1, points_f2, points_f3, points_f4 = custom_chaos_game_colored(iterations)

# Plot the attractor with different colors for each transformation
plt.figure(figsize=(2.5, 2.5))

# Set the bounds for the square frame
frame_size = 1.5
plt.xlim(-frame_size, frame_size)
plt.ylim(-frame_size, frame_size)

# Masculine and strong colors
plt.scatter(points_f1[:, 0], points_f1[:, 1], s=0.1, color='navy', label='f1')
plt.scatter(points_f2[:, 0], points_f2[:, 1], s=0.1, color='darkred', label='f2')
plt.scatter(points_f3[:, 0], points_f3[:, 1], s=0.1, color='darkgreen', label='f3')
plt.scatter(points_f4[:, 0], points_f4[:, 1], s=0.1, color='darkorange', label='f4')

plt.axis('equal')
plt.axis('off')
plt.title("Custom Attractor (Chaos Game with Colored Transformations)")
plt.legend()
plt.show()
```

We will show that this IFS satisfies the open set condition. Let us consider the set $A_0$ constructed by removing the upper right part of a square. Specifically, the vertices of $A_0$ are given by
$$ \{(0, 0), (1, 0), \left(1, \frac{\lambda}{1 - \lambda^2}\right), (0.5, 0.5), \left(\frac{\lambda}{1 - \lambda^2}, 1\right), (0, 1)\}. $$

It is straightforward to verify that for $0 < \lambda < \frac{1}{2}$ we have $f_i(A_0) \subset (A_0)$, for all $i = 1, 2, 3, 4$, and $f_i(A_0) \cap f_j(A_0) = \emptyset$ $(i \neq j)$, for all $i, j = 1, 2, 3, 4$. We define an open set $V$ to be the interior of $A_0$, which implies that
$$ f_i(V) \subset (V), \forall i = 1, \dots, 4; \quad f_i(V) \cap f_j(V) = \emptyset \ (i \neq j), \forall i, j = 1, \dots, 4. $$

Thus, the IFS $\{f_1, f_2, f_3, f_4\}$ satisfies the open set condition. Consequently, by Theorem 2.2, the Hausdorff dimension of corresponding self-similar set $A$, $\dim_H A$, is $s$.

Now let $A_j^i = f_i(A_{j-1}), \forall j, i = 1, 2, 3, 4$ and $A_j = \bigcup_{i=1}^4 f_i(A_{j-1})$ and denote $C, C_1, C_2, C_3, C_4$ be the convex hull of $A_0, A_1^1, A_2^2, A_3^3, A_4^4$, respectively.

Following a similar line of reasoning as in the proof of Proposition 1 in [1], it can be established that any straight line intersecting $A_0$ will also intersect $A_1$.

By induction, it follows that any straight line intersecting $A_0$ will intersect every $A_n$, and therefore the intersection $A = \bigcap_{n=1}^\infty A_n$ (by compactness). Notably, $A$ is the self-similar set generated by $\{f_1, f_2, f_3, f_4\}$, as documented in [4]. Moreover, since $f_i(A_0)$ are disjoint for all $i = 1, 2, 3, 4$, $A$ is totally disconnected.
